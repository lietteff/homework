```python
def count(data,n):
    lines=data.split('\n')
    k = [line.count(n) for line in lines]
    return k

data= '''Так старый хрыч, цыган Илья,
Глядит на удаль плясовую
Да чешет голову седую,
Под лад плечами шевеля…'''

n = 'а'
result = count(data, n)

for i, k in enumerate(result, start=1):
    print(f"В строке {i} символ '{n}' встречается {k} раз.")
```

```python
def decrypt(text):
    words = text.split()
    words_= [word[::-1] for word in words]
    text_=' '.join(words_)
    return text_

text = 'I evol nohtyp '
print(f'Расшифрованный текст:{decrypt(text)}')

```

```python
def write_to_file(file, numbers):
    with open(file, 'w') as file:
        for number in numbers:
            file.write(f"{number}\n")

def mult(file):
    product = 1
    with open(file, 'r') as file:
        for line in file:
            number = float(line.strip())
            product *= number
    return product

file = "numbers.txt"
numbers = [1,2,32,34,2.4,17]

write_to_file(file, numbers)

result = mult(file)
print(f"Произведение чисел: {result}")
```

```python
def count(file_name):
    letters = 0
    digits = 0

    with open(file_name, 'r', encoding='utf-8') as file:
        for line in file:
            for char in line:
                if char.isdigit():
                    digits += 1
                elif 'а' <= char.lower() <= 'я':
                    letters += 1

    return letters, digits

file = "real_numbers.txt"
letters, digits = count(file)
if letters > digits:
    print("В тексте больше русских букв.")
elif digits > letters:
    print("В тексте больше цифр.")
else:
    print("Количество русских букв и цифр одинаково.")

print(f"Русские буквы: {letters}, Цифры: {digits}")

"abc s as cse sc d  c d c x ф a 2 35 ы вп со 27 9 0 2 sh db ы й щ 2 3 ап 4"
```
