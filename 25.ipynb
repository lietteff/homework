```python
#-------------------------------------PROJECT-----------------------------------
from abc import ABC, abstractmethod
from typing import List

# Абстрактный класс Project
class Project(ABC):
    def __init__(self, name: str, start_date: str, end_date: str, budget: float, status: str, team: List[str]):
        self._name = name
        self._start_date = start_date
        self._end_date = end_date
        self._budget = budget
        self._status = status
        self._team = team

    @abstractmethod
    def calculate_progress(self):
        pass

    def __str__(self):
        return f"Название проекта: {self._name}, Статус: {self._status}"

    def __lt__(self, other):
        return self._budget < other._budget

    def __gt__(self, other):
        return self._budget > other._budget

# Подклассы для разных типов проектов
class SoftwareProject(Project):
    def __init__(self, name, start_date, end_date, budget, status, team, programming_languages):
        super().__init__(name, start_date, end_date, budget, status, team)
        self._programming_languages = programming_languages

    def calculate_progress(self):
        return "Прогресс: X% завершено"

    def __str__(self):
        return f"Программный проект: {self._name}, Языки: {', '.join(self._programming_languages)}"

class MarketingProject(Project):
    def __init__(self, name, start_date, end_date, budget, status, team, target_audience):
        super().__init__(name, start_date, end_date, budget, status, team)
        self._target_audience = target_audience

    def calculate_progress(self):
        return f"Охват аудитории: {self._target_audience} человек"

    def __str__(self):
        return f"Маркетинговый проект: {self._name}, Целевая аудитория: {self._target_audience}"

class ResearchProject(Project):
    def __init__(self, name, start_date, end_date, budget, status, team, research_field):
        super().__init__(name, start_date, end_date, budget, status, team)
        self._research_field = research_field

    def calculate_progress(self):
        return "Опубликовано Z статей"

    def __str__(self):
        return f"Исследовательский проект: {self._name}, Область исследований: {self._research_field}"

# Тестирование
software_project = SoftwareProject("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Андрей"], ["Python", "JavaScript"])
print(software_project)
print(software_project.calculate_progress())



# Интерфейс для отслеживания прогресса проекта
class Trackable(ABC):
    @abstractmethod
    def track_progress(self):
        pass

# Интерфейс для генерации отчетов по проекту
class Reportable(ABC):
    @abstractmethod
    def generate_report(self):
        pass

# Обновленные подклассы с интерфейсами
class SoftwareProject(Project, Trackable, Reportable):
    def __init__(self, name, start_date, end_date, budget, status, team, programming_languages):
        super().__init__(name, start_date, end_date, budget, status, team)
        self._programming_languages = programming_languages

    def calculate_progress(self):
        return "Прогресс: X% завершено"

    def track_progress(self):
        return f"Прогресс программного проекта {self._name}: X% завершено"

    def generate_report(self):
        return f"Отчет по проекту {self._name}: бюджет {self._budget}, статус {self._status}"

    def __str__(self):
        return f"Программный проект: {self._name}, Языки: {', '.join(self._programming_languages)}"

class MarketingProject(Project, Trackable, Reportable):
    def __init__(self, name, start_date, end_date, budget, status, team, target_audience):
        super().__init__(name, start_date, end_date, budget, status, team)
        self._target_audience = target_audience

    def calculate_progress(self):
        return f"Охват аудитории: {self._target_audience} человек"

    def track_progress(self):
        return f"Охват аудитории проекта {self._name}: {self._target_audience} человек"

    def generate_report(self):
        return f"Отчет по проекту {self._name}: бюджет {self._budget}, статус {self._status}, целевая аудитория {self._target_audience}"

    def __str__(self):
        return f"Маркетинговый проект: {self._name}, Целевая аудитория: {self._target_audience}"

class ResearchProject(Project, Trackable, Reportable):
    def __init__(self, name, start_date, end_date, budget, status, team, research_field):
        super().__init__(name, start_date, end_date, budget, status, team)
        self._research_field = research_field

    def calculate_progress(self):
        return "Опубликовано Z статей"

    def track_progress(self):
        return f"Опубликовано статей по проекту {self._name}: Z"

    def generate_report(self):
        return f"Отчет по проекту {self._name}: бюджет {self._budget}, статус {self._status}, область исследований {self._research_field}"

    def __str__(self):
        return f"Исследовательский проект: {self._name}, Область исследований: {self._research_field}"

# Тестирование
software_project = SoftwareProject("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Андрей"], ["Python", "JavaScript"])
print(software_project.track_progress())
print(software_project.generate_report())

# Обновленный класс проекта с методами сравнения
class ComparableProject(Project):
    def __eq__(self, other):
        return self._budget == other._budget and self._end_date == other._end_date

    def __lt__(self, other):
        return (self._budget, self._end_date) < (other._budget, other._end_date)

    def __gt__(self, other):
        return (self._budget, self._end_date) > (other._budget, other._end_date)

# Тестирование методов сравнения
project1 = ComparableProject("Проект A", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Аиша"])
project2 = ComparableProject("Проект B", "2025-06-01", "2025-12-01", 60000, "Запущен", ["Елизавета", "Аиша"])
project3 = ComparableProject("Проект C", "2025-07-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Аиша"])

print(f"Проект 1 == Проект 3: {project1 == project3}")  # Одинаковый бюджет и срок - True
print(f"Проект 1 < Проект 2: {project1 < project2}")    # Меньший бюджет и срок - True
print(f"Проект 2 > Проект 3: {project2 > project3}")    # Больший бюджет - True
```

```python
#-----------------------------------TEAM----------------------------------------
# Класс для управления командой
from models.project import Project
class Employee:
    def __init__(self, name: str, position: str):
        self._name = name
        self._position = position

    def __str__(self):
        return f"{self._name} ({self._position})"

class Team:
    def __init__(self):
        self._members = []

    def add_member(self, employee: Employee):
        self._members.append(employee)

    def remove_member(self, employee: Employee):
        self._members.remove(employee)

    def get_members(self):
        return [str(member) for member in self._members]

    def get_team_size(self):
        return len(self._members)

# Композиция: Добавление лидера проекта
class ProjectWithLeader(Project):
    def __init__(self, name, start_date, end_date, budget, status, team, leader: Employee):
        super().__init__(name, start_date, end_date, budget, status, team)
        self._leader = leader

    def __str__(self):
        return f"Название проекта: {self._name}, Лидер: {self._leader}, Статус: {self._status}"

# Тестирование
team = Team()
leader = Employee("Ислам", "Руководитель проекта")
team.add_member(Employee("Аиша", "Разработчик"))
team.add_member(Employee("Лиза", "Дизайнер"))

project = ProjectWithLeader("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", team.get_members(), leader)

print(project)
print("Команда проекта:", team.get_members())
print("Количество участников:", team.get_team_size())
```

```python
#--------------------------------PROGRESS---------------------------------------
from abc import ABC, abstractmethod
from models.project import Project,SoftwareProject,MarketingProject

# Базовый класс для расчета прогресса
class ProgressCalculator(ABC):
    def calculate_progress(self, project):
        self._validate_project(project)
        current_state = self._get_current_state(project)
        progress = self._compute_progress(current_state)
        return f"Прогресс проекта {project._name}: {progress}% завершено"

    @abstractmethod
    def _get_current_state(self, project):
        pass

    @abstractmethod
    def _compute_progress(self, state):
        pass

    def _validate_project(self, project):
        if not isinstance(project, Project):
            raise ValueError("Переданный объект не является проектом.")

# Подклассы для разных типов проектов
class SoftwareProgressCalculator(ProgressCalculator):
    def _get_current_state(self, project):
        return len(project._programming_languages) * 10  # Пример расчета на основе количества языков

    def _compute_progress(self, state):
        return min(state, 100)  # Прогресс не может превышать 100%

class MarketingProgressCalculator(ProgressCalculator):
    def _get_current_state(self, project):
        return project._target_audience / 1000  # Пример расчета на основе охвата аудитории

    def _compute_progress(self, state):
        return min(state, 100)

# Тестирование шаблонного метода
software_project = SoftwareProject("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Аиша"], ["Python", "JavaScript"])
marketing_project = MarketingProject("Рекламная кампания", "2025-05-01", "2025-09-01", 30000, "Запущен", ["Ислам", "Лиза"], 50000)

software_progress_calculator = SoftwareProgressCalculator()
marketing_progress_calculator = MarketingProgressCalculator()

print(software_progress_calculator.calculate_progress(software_project))
print(marketing_progress_calculator.calculate_progress(marketing_project))
```

```python
#----------------------------DECORATOR------------------------------------------
# Исключение для нарушения прав доступа
from models.project import SoftwareProject
class PermissionDeniedError(Exception):
    pass

# Декоратор проверки прав доступа
def check_permissions(required_role):
    def decorator(func):
        def wrapper(self, *args, **kwargs):
            if not hasattr(self, "_user_role") or self._user_role != required_role:
                raise PermissionDeniedError(f"Доступ запрещен: требуется роль {required_role}")
            return func(self, *args, **kwargs)
        return wrapper
    return decorator

# Обновленный класс проекта с проверкой прав
class SecureSoftwareProject(SoftwareProject):
    def __init__(self, name, start_date, end_date, budget, status, team, programming_languages, user_role):
        super().__init__(name, start_date, end_date, budget, status, team, programming_languages)
        self._user_role = user_role

    @check_permissions("admin")
    def generate_report(self):
        report = super().generate_report()
        return f"(Отчет создан администратором) {report}"

# Тестирование декоратора
admin_project = SecureSoftwareProject("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Аиша"], ["Python", "JavaScript"], "admin")
user_project = SecureSoftwareProject("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Аиша"], ["Python", "JavaScript"], "user")

print(admin_project.generate_report())  # Успешное выполнение
try:
    print(user_project.generate_report())  # Ошибка прав доступа
except PermissionDeniedError as e:
    print(e)
```

```python
#-----------------------------------DOCS----------------------------------------
import logging
from abc import ABC, abstractmethod
from typing import List

logging.basicConfig(filename="project.log", level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

class Project(ABC):
    """
    Абстрактный класс, представляющий проект.
    """

    def __init__(self, name: str, start_date: str, end_date: str, budget: float, status: str, team: List[str]):
        """
        Инициализация проекта.

        :param name: Название проекта.
        :param start_date: Дата начала.
        :param end_date: Дата окончания.
        :param budget: Бюджет проекта.
        :param status: Статус проекта.
        :param team: Список участников проекта.
        """
        self._name = name
        self._start_date = start_date
        self._end_date = end_date
        self._budget = budget
        self._status = status
        self._team = team

    @abstractmethod
    def calculate_progress(self):
        """Абстрактный метод для расчета прогресса проекта."""
        pass

    def __str__(self):
        """
        Возвращает строковое представление проекта.

        :return: Строка с названием и статусом проекта.
        """
        return f"Название проекта: {self._name}, Статус: {self._status}"

class LoggingMixin:
    """
    Миксин для логирования действий с проектами.
    """

    def log_action(self, action: str):
        """
        Записывает действие, связанное с проектом, в лог.

        :param action: Описание действия.
        """
        logging.info(f"Проект '{self._name}': {action}")

class SoftwareProject(Project, LoggingMixin):
    """
    Класс, представляющий программный проект.
    """

    def __init__(self, name, start_date, end_date, budget, status, team, programming_languages):
        """
        Инициализация программного проекта.

        :param programming_languages: Список используемых языков программирования.
        """
        super().__init__(name, start_date, end_date, budget, status, team)
        self._programming_languages = programming_languages

    def calculate_progress(self):
        """
        Рассчитывает прогресс проекта.

        :return: Строка с текущим прогрессом.
        """
        return "Прогресс: X% завершено"

    def track_progress(self):
        """
        Отслеживает прогресс проекта.

        :return: Строка с текущими данными о прогрессе.
        """
        progress = f"Прогресс программного проекта {self._name}: X% завершено"
        self.log_action("Обновление прогресса")
        return progress

    def generate_report(self):
        """
        Генерирует отчет по проекту.

        :return: Строка с данными отчета.
        """
        report = f"Отчет по проекту {self._name}: бюджет {self._budget}, статус {self._status}"
        self.log_action("Генерация отчета")
        return report

# Тестирование документации и логирования
project = SoftwareProject("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Аиша"], ["Python", "JavaScript"])
print(project.track_progress())
print(project.generate_report())
```

```python
#-------------------------------METACLASS---------------------------------------
# Метакласс для регистрации подклассов
from abc import ABC, abstractmethod
from models.project import Project
from typing import List

class ProjectMeta(type):
    _registry = {}

    def __init__(cls, name, bases, class_dict):
        super().__init__(name, bases, class_dict)
        if cls.__name__ != "Project":
            ProjectMeta._registry[cls.__name__] = cls

    @classmethod
    def get_project_class(cls, project_type):
        return cls._registry.get(project_type)

# Обновляем базовый класс с метаклассом
class Project(ABC, metaclass=ProjectMeta):
    def __init__(self, name: str, start_date: str, end_date: str, budget: float, status: str, team: List[str]):
        self._name = name
        self._start_date = start_date
        self._end_date = end_date
        self._budget = budget
        self._status = status
        self._team = team

    @abstractmethod
    def calculate_progress(self):
        pass

    def __str__(self):
        return f"Название проекта: {self._name}, Статус: {self._status}"

# Тестирование метакласса
print(ProjectMeta._registry)  # Показывает, какие подклассы зарегистрированы
software_project_class = ProjectMeta.get_project_class("SoftwareProject")
if software_project_class:
    instance = software_project_class("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Аиша"], ["Python", "JavaScript"])
    print(instance)
```

```python
#------------------------------SER&DESER----------------------------------------
import json
from models.project import Project

# Добавляем методы сериализации и десериализации в Project
class SerializableProject(Project):
    def to_dict(self):
        return {
            "name": self._name,
            "start_date": self._start_date,
            "end_date": self._end_date,
            "budget": self._budget,
            "status": self._status,
            "team": self._team
        }

    @classmethod
    def from_dict(cls, data):
        return cls(data["name"], data["start_date"], data["end_date"], data["budget"], data["status"], data["team"])

    def save_to_file(self, filename):
        with open(filename, "w") as file:
            json.dump(self.to_dict(), file, indent=4)

    @classmethod
    def load_from_file(cls, filename):
        with open(filename, "r") as file:
            data = json.load(file)
        return cls.from_dict(data)

# Тестирование сериализации
project = SerializableProject("Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Аиша"])
project.save_to_file("project.json")

# Тестирование десериализации
loaded_project = SerializableProject.load_from_file("project.json")
print(loaded_project)
```

```python
#------------------------------CHAIN--------------------------------------------
# Паттерн "Цепочка обязанностей"
from abc import ABC,abstractmethod
class BudgetApprovalHandler(ABC):
    def __init__(self, successor=None):
        self._successor = successor

    @abstractmethod
    def approve_budget(self, amount):
        pass

# Руководитель проекта может одобрить до 1000 единиц
class ProjectLeader(BudgetApprovalHandler):
    def approve_budget(self, amount):
        if amount <= 1000:
            return f"Руководитель проекта одобрил бюджетное изменение на {amount} единиц."
        elif self._successor:
            return self._successor.approve_budget(amount)
        return "Запрос отклонен."

# Финансовый отдел может одобрить до 5000 единиц
class FinanceDepartment(BudgetApprovalHandler):
    def approve_budget(self, amount):
        if amount <= 5000:
            return f"Финансовый отдел одобрил бюджетное изменение на {amount} единиц."
        elif self._successor:
            return self._successor.approve_budget(amount)
        return "Запрос отклонен."

# Директор может одобрить любое изменение бюджета
class Director(BudgetApprovalHandler):
    def approve_budget(self, amount):
        return f"Директор одобрил бюджетное изменение на {amount} единиц."

# Формирование цепочки обработчиков
project_leader = ProjectLeader(FinanceDepartment(Director()))

# Тестирование цепочки обязанностей
print(project_leader.approve_budget(800))   # Руководитель одобрит
print(project_leader.approve_budget(3000))  # Финансовый отдел одобрит
print(project_leader.approve_budget(10000)) # Директор одобрит
```

```python
#-------------------------------FABRIC------------------------------------------
# Класс для фабричного метода
from models.project import Project,SoftwareProject,MarketingProject,ResearchProject
from typing import List
class ProjectFactory:
    @staticmethod
    def create_project(project_type: str, name: str, start_date: str, end_date: str, budget: float, status: str,
                       team: List[str], extra_data):
        project_classes = {
            "software": SoftwareProject,
            "marketing": MarketingProject,
            "research": ResearchProject
        }

        project_class = project_classes.get(project_type.lower())
        if not project_class:
            raise ValueError(f"Неизвестный тип проекта: {project_type}")

        return project_class(name, start_date, end_date, budget, status, team, extra_data)


# Тестирование фабричного метода
software_project = ProjectFactory.create_project(
    "software", "Разработка веб-приложения", "2025-04-01", "2025-10-01", 50000, "В процессе", ["Елизавета", "Андрей"],
    ["Python", "JavaScript"]
)

marketing_project = ProjectFactory.create_project(
    "marketing", "Рекламная кампания", "2025-05-01", "2025-09-01", 30000, "Запущен", ["Мария", "Дмитрий"], 50000
)

print(software_project)
print(marketing_project)
```
