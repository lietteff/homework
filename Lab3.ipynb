```python
#1
def sum_mult(arr):
    a = arr[:5]
    return sum(a), (lambda x: eval("*".join(map(str, x))))(a)

array = [1, 2, 3, 4, 5, 6, 7]
sm, mult = sum_mult(array)
print(f"Сумма первых пяти элементов: {sm}")
print(f"Произведение первых пяти элементов: {mult}")
```

```python
#2
def generate(n):
    return [2**i for i in range(1, n + 1)]

N = 5
result = generate(N)
print(f"Массив степеней двойки: {result}")
```

```python
#3
def generate(A):
    B = []
    sm = 0
    for element in A:
        sm += element
        B.append(sm)
    return B

A = [1, 2, 3, 4, 5]
B = generate(A)
print(f"Новый массив B: {B}")
```

```python
def sum_mult(matrix):
    sm = 0
    mult = 1
    has_even = False

    for row in matrix:
        for element in row:
            if element % 2 == 0:
                has_even = True
                sm += element
                mult *= element

    if not has_even:
        mult = 0

    return sm, mult

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

sm, mult = sum_mult(matrix)
print(f"Сумма четных элементов: {sm}")
print(f"Произведение четных элементов: {mult}")
```

```python
def filter(data, porog):
    return {key: value for key, value in data.items() if value > porog}

dict = {"a": 10,"b": 5,"c": 20,"d": 3}

porog = 7
filtered_dict = filter(dict, porog)
print("Отфильтрованный словарь:", filtered_dict)
```

```python
def key_max(input):
    if not input:
        return None
    max_key = max(input, key=input.get)
    return max_key

data = {'a': 1, 'b': 586, 'c': 157, 'd': 855}
result = key_max(data)
print(f"Ключ с максимальным значением: {result}")
```

```python
def key_min(input):
    if not input:
        return None
    min_key = min(input, key=input.get)
    return min_key

data = {'a': 1, 'b': 586, 'c': 157, 'd': 855}
result = key_min(data)
print(f"Ключ с минимальным значением: {result}")
```

```python
def count(data):
    k = {}
    for value in data.values():
        if value in k:
            k[value] += 1
        else:
            k[value] = 1
    return k

dict = {"a": 3,"b": 3,"c": 7,"d": 5,"e": 3,"f":7}

value = count(dict)
print("Частота значений:", value)
```

```python
def is_set(set1, set2):
    return set1.issubset(set2)

set_a = {1, 2, 3}
set_b = {1, 2, 3, 4, 5}
result = is_set(set_a, set_b)
if result:
    print("Первое множество является подмножеством второго.")
else:
    print("Первое множество не является подмножеством второго.")
```

```python
import random

def generate(n, start, end):
    if n > (end - start + 1):
        raise ValueError
    A = set()
    while len(A) < n:
        A.add(random.randint(start, end))
    return A

n = 5
start = 1
end = 10
A = generate(n, start, end)
print(f"Сгенерированное множество: {A}")
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def prime_tuple(start, end):
    return tuple(num for num in range(start, end + 1) if is_prime(num))

start = 1
end = 51
result = prime_tuple(start, end)
print(f"Кортеж простых чисел: {result}")
```
