```python
from math import sqrt
x1=float(input('x1='))
y1=float(input('y1='))
x2=float(input('x2='))
y2=float(input('y2='))
print(f'Координаты точки А=({x1},{y1})')
print(f'Координаты точки B=({x2},{y2})')

def f(x1,y1,x2,y2):
    if x1==x2 and y1==y2:return 0
    if sqrt(x1**2+y1**2)>sqrt(x2**2+y2**2): return 1
    if sqrt(x1 ** 2 + y1 ** 2) < sqrt(x2 ** 2 + y2 ** 2): return 2
if f(x1,y1,x2,y2)==0:print('Точки совпадают')
if f(x1,y1,x2,y2)==1:print('Точка В ближе')
if f(x1,y1,x2,y2)==2:print('Точка А ближе')
```

#7. Последовательности: группа Series. Вложенные циклы


image.png

```python
def total_distance(first_day, days, percent):
    total = 0
    daily_distance = first_day

    for day in range(1, days + 1):
        total += daily_distance
        daily_distance += daily_distance * (percent / 100)

    return total

first_day_distance = 10
number_of_days = 7
percent = 10

result = total_distance(first_day_distance, number_of_days, percent)
print(f"Суммарный путь за {number_of_days} дней: {result:.2f} км")
```
