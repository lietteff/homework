#2. Наследование

image.png

```python
class Document:
    def __init__(self, name, author, content):
        self.name = name
        self.author = author
        self.content = content

    def display_info(self):
        return f"Название: {self.name}\nАвтор: {self.author}\nСодержимое: {self.content}"


class Recipe(Document):
    def __init__(self, name, author, content, ingredients, cook_time):
        super().__init__(name, author, content)
        self.ingredients = ingredients
        self.cook_time = cook_time

    def update_cook_time(self, new_cook_time):
        self.cook_time = new_cook_time

    def display_info(self):
        basic_info = super().display_info()
        return f"{basic_info}\nИнгредиенты: {', '.join(self.ingredients)}\nВремя приготовления: {self.cook_time} минут"

recipe = Recipe(
    "Печенье",
    "Елизавета",
    "Имбирное печенье от бабушки.",
    ["сода", "какао", "имбирь", "мед", "мука"],
    60
)

print(recipe.display_info())
recipe.update_cook_time(75)
print(recipe.display_info())
```

#3. Инкапсуляция

```python
class Recipe:
    def __init__(self, name, ingredients, cooking_time):
        self.__name = name
        self.__ingredients = ingredients
        self.__cooking_time = cooking_time
    def get_name(self):
        return self.__name

    def get_ingredients(self):
        return self.__ingredients

    def get_cooking_time(self):
        return self.__cooking_time

    def set_name(self, new_name):
        self.__name = new_name

    def set_ingredients(self, new_ingredients):
            self.__ingredients = new_ingredients

    def set_cooking_time(self, new_time):
        if isinstance(new_time, int) and new_time > 0:
            self.__cooking_time = new_time
        else:
            print("Ошибка")

    def display_recipe(self):
        return (f"Название: {self.__name}\n"
                f"Ингредиенты: {', '.join(self.__ingredients)}\n"
                f"Время приготовления: {self.__cooking_time} минут")

recipe = Recipe("Печенье", ["сода", "мед", "какао"], 60)
print(recipe.display_recipe())

recipe.set_cooking_time(90)
recipe.set_ingredients(["мука", "сода", "какао", "сахар","мед"])
print(recipe.display_recipe())
```
